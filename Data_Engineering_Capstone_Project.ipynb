Task 5 : Use Class And Function :
Define a Python class RetailSalesAnalyzer
with methods for each of the above tasks (loading data,
calculating statistics, and creating visualizations)



Task 6 : Write a script that creates an instance of
RetailSalesAnalyzer, calls its methods,
and displays the results

```python
import pandas as pd
import matplotlib.pyplot as plt
class RetailSalesAnalyzer:
  def __init__(self):
    self.data = pd.read_csv('/content/retail_sales.csv') #load the csv file to pandas data frame
    self.data['Data'] = pd.to_datetime(self.data['Date']) #Correct the date column format

  def data_clean(self):
    self.data.dropna(inplace=True) #Remove the rows that has empty cells in any column

  def total_sales_per_product(self):
    return self.data.groupby('Product')['Sales'].sum() #Get sales per product and return sum

  def best_selling_product(self):
    return self.total_sales_per_product().sort_values(ascending=False).index[0]#Call the total sales per product and sort them in Descending order

  def average_daily_sales(self):
    return self.data['Sales'].mean() #get sales and return average

  def plot_sales_trends(self):
    self.total_sales_per_product().plot(kind='line')
    plt.title('Sales Trends Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.show()

  def plot_sales_per_product(self):
    self.total_sales_per_product().plot(kind='bar')
    plt.title('Sales Per Product')
    plt.xlabel('Product')
    plt.ylabel('Sales')
    plt.show()
```

```python
retailSalesAnalyzer = RetailSalesAnalyzer()
print("Cleaning Data Done", retailSalesAnalyzer.data_clean())
print("Total Sales Per Product", retailSalesAnalyzer.total_sales_per_product())
print("Best Selling Product", retailSalesAnalyzer.best_selling_product())
print("Average Daily Sales", retailSalesAnalyzer.average_daily_sales())
print("Line Plot\n")
retailSalesAnalyzer.plot_sales_trends()
print("Bar Chart\n")
retailSalesAnalyzer.plot_sales_per_product()
```
